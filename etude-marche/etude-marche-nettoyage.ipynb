```python
df_population.drop(['Code Domaine', 'Code Élément', 'Code Produit', 'Code année', 'Symbole', 'Description du Symbole', 'Note'], axis=1, inplace=True)
```

```python
df_population = df_population[df_population['Année'] == 2017]
```

```python
def get_df_name(df):
    name =[x for x in globals() if globals()[x] is df][0]
    return name

for i in [df_fao_data, df_disponibilite_alimentaire, df_population]:
    print("Nombre de lignes en doublons pour le dataframe {}: {}".format(get_df_name(i), len(i[i.duplicated()])))
```

```python
df_concat = pd.concat([df_fao_data, df_disponibilite_alimentaire, df_population])
```

```python
df_final = df_concat.pivot_table(index='Zone', values='Valeur', columns=['Élément', 'Produit', 'Année']).reset_index()
df_final.info()
```

```python
df_final[(df_final[('Exportations - Quantité', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Importations - Quantité', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Production', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Disponibilité alimentaire (Kcal/personne/jour)', 'Viande de Volailles', 2017)].isna())]
```

```python
df_final[(df_final[('Valeur US $, aux prix du 2015', 'Produit Intérieur Brut', 2017)].isna())]
```

```python
# Province de Taiwan est un élément important mais le PIB est introuvable dans le dataset de FAO
# On renseigne le PIB manuellement après des recherches sur le Web
df_final.loc[df_final['Zone'] == 'Chine, Taiwan Province de', [('Valeur US $, aux prix du 2015', 'Produit Intérieur Brut', 2017)]] = 590780
```

```python
df_final[df_final['Zone'] == 'Chine, Taiwan Province de']
```

```python
lst_pays_to_drop = np.array(df_final[(df_final[('Exportations - Quantité', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Importations - Quantité', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Production', 'Viande de Volailles', 2017)].isna()) &
(df_final[('Disponibilité alimentaire (Kcal/personne/jour)', 'Viande de Volailles', 2017)].isna())]['Zone'])
print("Liste des pays à supprimer car aucune donnée d'export/import et production de viande de volailles: \n{}".format(lst_pays_to_drop))
```

```python
lst_pays_to_drop = np.array(df_final[(df_final[('Exportations - Quantité', 'Viande de Volailles', 2017)].isna()) & (df_final[('Importations - Quantité', 'Viande de Volailles', 2017)].isna()) & (df_final[('Production', 'Viande de Volailles', 2017)].isna())]['Zone'])
df_datas_for_cah = df_final[~df_final['Zone'].isin(lst_pays_to_drop)]
population_millions = pd.DataFrame(df_datas_for_cah[('Population totale', 'Population-Estimations', 2017)] / 1000)
ppa = pd.DataFrame((df_datas_for_cah[('Valeur US $, aux prix du 2015', 'Produit Intérieur Brut', 2017)] / (df_datas_for_cah[('Population totale', 'Population-Estimations', 2017)] / 1000)))
population_millions.columns = pd.MultiIndex.from_tuples([('Population totale millions', 'Population-Estimations', 2017)])
ppa.columns = pd.MultiIndex.from_tuples([('Valeur US $, aux prix du 2015','Produit Intérieur Brut Par Habitant', 2017)])
df_merge = pd.merge(df_datas_for_cah, population_millions, left_index=True, right_index=True)
df_datas_for_cah = pd.merge(df_merge, ppa, left_index=True, right_index=True)
df_datas_for_cah.drop([('Valeur US $, aux prix du 2015','Produit Intérieur Brut', 2017), ('Population totale', 'Population-Estimations', 2017)], axis=1, inplace=True)
```

```python
#On supprime les pays trop importants pour se focaliser sur un périmètre plus accessible et moins concurrentiel
lst_pays_excentrés = ("États-Unis d'Amérique", "Brésil", "Chine, continentale", "Inde")
df_datas_for_cah = df_datas_for_cah.replace([np.inf, -np.inf], np.nan).apply(lambda x: x.fillna(x.min()),axis=0).reset_index().sort_index(axis=1).drop(['index'], axis=1)
df_datas_for_cah.drop(df_datas_for_cah[df_datas_for_cah['Zone'].isin(lst_pays_excentrés)].index, inplace=True)
df_datas_for_cah.info()
```
